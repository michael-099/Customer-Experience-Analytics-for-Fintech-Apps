```python
import pandas as pd
from google_play_scraper import reviews
import time
```

```python
def scrape(id, rows):
    results, continuation_token = reviews(
        id,
        lang="en",
        country="et",
        count=rows,
    )
    print(results)
    return results
```

```python
CBE_data = scrape("com.combanketh.mobilebanking", 500)
```

```python
BOA_data = scrape("com.boa.boaMobileBanking", 500)
```

```python
DB_data = scrape("com.dashen.dashensuperapp", 500)
```

```python
print(CBE_data[0].keys() == BOA_data[0].keys() == DB_data[0].keys())
```

```python
df_CBE = pd.DataFrame(CBE_data)
df_BOA = pd.DataFrame(BOA_data)
df_DB = pd.DataFrame(DB_data)

df_CBE["bank"] = "CBE"
df_BOA["bank"] = "BOA"
df_DB["bank"] = "DB"
```

```python
df = pd.concat([df_CBE, df_BOA, df_DB], ignore_index=True)
```

```python
df["source"] = "google play"
```

```python
df = df.rename(
    columns={
        "content": "review",
        "score": "rating",
        "at": "date",
    }
)
```

```python
df = df[["review", "rating", "date", "bank", "source"]]
```
